```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='white', color_codes=True)
```

```python
#loading the data
df = pd.read_csv('vgsales.csv')
```

```python
#what is the most common platform?
df['Platform'].mode()[0]
```

```python
#what is the most common Publisher?
df['Publisher'].mode()[0]
```

```python
#What is the North American Sale total?
df['NA_Sales'].sum()
```

```python
#What is the Europe Sale total?
df['EU_Sales'].sum()
```

```python
#What is the Other Sale total?
df['Other_Sales'].sum()
```

```python
#What is the Global Sale total?
df['Global_Sales'].sum()
```

```python
#what is the top 10 highest grossing games?
df.nlargest(n=10, columns=['Global_Sales'])
```

```python
#top 5 most sold game in North America?
top_na = df.sort_values(by=['NA_Sales'], ascending=False)
top_na.head(5)
```

```python
#top 5 most sold game in Europe?
top_eu = df.sort_values(by=['EU_Sales'], ascending=False)
top_eu.head(5)
```

```python
#What game sold the least?
least_sold = df[['Name','Global_Sales']].min()
least_sold
```

```python
df['Publisher'].nunique()
```

```python
#top 5 sega games
sg = df[df['Publisher'] == 'Sega']
sg.sort_values(by = ["Global_Sales"], ascending = False).head(5)
```
